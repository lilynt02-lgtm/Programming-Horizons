```python
#Add new columns
df['Year of joining'] = [2016, 2017,2018,2019,2020,2021,2022,2023]
df.head()
#there are 8 values (years) because there are 8 rows
```

```python
#create/add an other column 'Bonus' correspond 10% of the Salary
df['Bonus'] = df['Salary']*0.10
df.head()
```

```python
#Remove a column = drop()
df.drop('Year of joining', axis = 1, inplace = True)
#because all columns are along X aqual to 1. True means make the changes permanent in the dataframe
df.head()
```

```python
#Statistical operations
#Count : calculate the total
print("Total Employees : ", df['Name'].count())

#mean() : calculate the average
print("Average Salary : ", df['Salary'].mean())

#value_counts() : count the frequency of unique value a particular column
print("Department-wise Employees Count : ", df['Department'].value_counts()) #how many times a particular department appears
```

```python
#Group the data : Groupby()
print("Department-wise average alaries : ",
      df.groupby('Department')['Salary'].mean())

 #the average salary of each department
```

```python
#Handle missing values
print(df)
```

```python
#Check for the missing values in each columns
print(df.isnull().sum())
```

```python
#dropping the rows with missing values
df2 = df.dropna(axis = 0, inplace = True)
print(df2)
#dropna means remove all rows from the original dataframe
#axis = 1 means column, axis = 0 means row
#inplace = True means the changes permanent, = False means the changes tempory
```
